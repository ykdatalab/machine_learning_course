```python
def mile(x) :
    if x['fuel']  == 'Petrol' :
        return x['mileage'] / 80.43
    
    elif x['fuel'] == 'Diesel' :
        return x['mileage'] / 73.56
    
    elif x['fuel'] == 'LPG' :
        return x['mileage'] / 40.85
    
    else :
        return x['mileage'] /44.23
```

```python
car['mileage'] = car.apply(mile, axis=1)
```

```python
car.drop('mileage_unit', axis=1, inplace=True)
```

```python
def torque_unit(x) :
    if 'NM' in str(x) :
        return 'Nm'
    
    elif 'KGM' in str(x) :
        return 'kgm'
```

```python
car['torque_unit'] = car['torque'].apply(torque_unit)
car['torque_unit'].unique()
```

```python
def split_num(x) :
    x = str(x)
    for i, j in enumerate(x) :
        if j not in '0123456789. ' :
            cut = i
            break
    return x[:cut]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(car.drop(
    'selling_price', axis=1), car['selling_price'], test_size=0.2, random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=100)

model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

print('train_rmse:', mean_squared_error(y_train, train_pred) ** 0.5,
      'test_rmse:', mean_squared_error(y_test, test_pred) ** 0.5)
```

```python
from sklearn.model_selection import KFold

car
```

```python
car.reset_index(drop=True, inplace=True)
car
```

```python
kf = KFold(n_splits=5)

X = car.drop('selling_price', axis=1)
y = car['selling_price']
```

```python
for i, j in kf.split(X) :
    print(i, j)
```

```python
for train_index, test_index in kf.split(X) :
    
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
train_rmse_total = []
test_rmse_total = []

for train_index, test_index in kf.split(X) :
    X_train, X_test, = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    model = RandomForestRegressor(random_state=100)
    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    train_rmse = mean_squared_error(y_train, train_pred) ** 0.5
    test_rmse = mean_squared_error(y_test, test_pred) ** 0.5
    
    train_rmse_total.append(train_rmse)
    test_rmse_total.append(test_rmse)
    
    train_rmse_total
```

```python
print('train_rmse: ', sum(train_rmse_total)/ 5, 
      'test_rmse: ', sum(test_rmse_total) /5)
```

```python
train_rmse_total = []
test_rmse_total = []

for train_index, test_index in kf.split(X) :
    
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y[train_index], y[test_index]
    
    model = RandomForestRegressor(n_estimators= 300, 
                                  max_depth=50, 
                                  min_samples_split=5,
                                  min_samples_leaf=1, n_jobs= -1,
                                  random_state=100)
    
    model.fit(X_train, y_train)
    
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    train_rmse = mean_squared_error(y_train, train_pred) ** 0.5
    test_rmse = mean_squared_error(y_test, test_pred) ** 0.5
    
    train_rmse_total.append(train_rmse)
    test_rmse_total.append(test_rmse)
    
    print('train_rmse: ', sum(train_rmse_total) / 5, 
          "test_rmse: ", sum(test_rmse_total) / 5)
```
